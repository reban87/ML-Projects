```python
import pandas as pd
df=pd.read_csv("us-embassy, kathmandu-air-quality .csv")

print(df)
```

```python
df.columns
```

```python
#last three months AQI
df['date'] = pd.to_datetime(df.date)
mask = (df['date'] >='2020-03-01') & (df['date']  <= '2020-06-01')

dfl = df.loc[mask]
dfl
```

```python
import matplotlib.pyplot as plt

dates = dfl['date']
pm25 = dfl['PM25']
pm25 = [int(i) for i in pm25]

plt.figure(figsize=(10,8))

length = len(dates)

plt.plot(dates,pm25,color='green',label='March to June PM2.5 values')
plt.title('PM2.5 values from March to June of Kathmandu- US Embassy Station AQI Data')
plt.xlabel('Dates of Last March to June')
plt.ylabel('PM2.5 values')
plt.legend()
plt.show()
```
